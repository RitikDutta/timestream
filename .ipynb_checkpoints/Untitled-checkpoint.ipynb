```python
import os
from flask import Flask, request, render_template, jsonify
from youtube_transcript_api import YouTubeTranscriptApi, NoTranscriptFound
import google.generativeai as genai
from dotenv import load_dotenv
import re
class Timestream:
    def __init__(self, api_key='678df2f442c5b90111351c2f'):
        """
        Initializes the YouTubeTranscriptExtractor with the necessary API credentials.

        Parameters:
            api_key (str): The API key for authorization.
        """
        self.url = "https://www.youtube-transcript.io/api/transcripts"
        self.headers = {
            "Authorization": f"Basic {api_key}",
            "Content-Type": "application/json"
        }

    def format_time(self, seconds):
        hours = int(seconds // 3600)
        minutes = int((seconds % 3600) // 60)
        seconds = int(seconds % 60)
        if hours > 0:
            return f"{hours:02}:{minutes:02}:{seconds:02}"
        else:
            return f"{minutes:02}:{seconds:02}"

    def get_youtube_video_id(self, url):
        # Regex patterns for different YouTube URL formats
        patterns = [
            r'(?:https?:\/\/)?(?:www\.)?(?:youtube\.com\/(?:[^\/\n\s]+\/\S+\/|(?:v|e(?:mbed)?)\/|\S*?[?&]v=)|youtu\.be\/)([a-zA-Z0-9_-]{11})',
        ]

        for pattern in patterns:
            match = re.match(pattern, url)
            if match:
                return match.group(1)
        return None

    def get_instructions(self, instructions):
        with open(instructions, 'r') as file:
            instructions = file.read()
        return instructions
        
    def get_youtube_transcript(self, video_id):
        transcript_text = []
        languages = ['en', 'hi']
        print(f"----------{video_id}----------")
        for language in languages:
            try:
                transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=[language])
                print(type(transcript))
                print(transcript)
                print(f"Transcript found in language: {language}")
                for entry in transcript:
                    start_time = self.format_time(entry['start'])
                    text = entry['text']
                    transcript_text.append(f"{start_time} {text}")
                return transcript_text  # Return if transcript is found
            except NoTranscriptFound as e:
                print(f"Transcript not found for language: {language}")
        
        # If no transcript is found for any language, raise the exception with required arguments
        raise NoTranscriptFound(requested_language_codes=languages, transcript_data=None)

    #as youtube have blocked the api from ipv6 commonly used by cloud platforms so we did a different technique to get the transcript
    def get_youtube_transcript_different_api(self, video_id, preferred_language='English'):
        
        response_tr2 = self._make_api_request(video_id)
        if not response_tr2:
            print("Failed to retrieve transcript data.")
            return []

        # Step 2: Extract 'tracks' from the response
        try:
            tracks = response_tr2[0]['tracks']
        except (IndexError, KeyError) as e:
            print(f"Error extracting 'tracks' from response: {e}")
            return []

        # Step 3: Extract preferred language transcript
        transcript = self._extract_transcript_with_floats(tracks, preferred_language)

        # Step 4: If preferred language not found, fallback to alternative language
        if not transcript:
            fallback_language = 'Hindi (auto-generated)' if preferred_language.lower() == 'english' else 'English'
            print(f"Preferred language '{preferred_language}' not found. Falling back to '{fallback_language}'.")
            transcript = self._extract_transcript_with_floats(tracks, fallback_language)

            if not transcript:
                print(f"No transcript available for '{preferred_language}' or '{fallback_language}'.")
                return []

        # Step 5: Format the transcript entries
        transcript_text = []
        for entry in transcript:
            start_time = self._format_time(entry['start'])
            text = entry['text']
            transcript_text.append(f"{start_time} {text}")

        return transcript_text

    def _make_api_request(self, video_id):
        url = "https://www.youtube-transcript.io/api/transcripts"
        headers = {
            "Authorization": "Basic 678df2f442c5b90111351c2f",
            "Content-Type": "application/json"
        }

        # Define the payload
        payload = {
            "ids": ["i6bTLmBYNf4"]
        }

        # Make the POST request
        response = requests.post(url, headers=headers, json=payload)

        # Check if the request was successful
        if response.status_code == 200:
            response_tr2 = response.json()  # Parse the JSON response
            print("Response successfully received and parsed!")
            print(json.dumps(response_tr, indent=2))  # Pretty-print JSON response
        else:
            print(f"Request failed with status code {response.status_code}: {response.text}")
        return response_tr2

    def _extract_transcript_with_floats(self, tracks, language):
        """
        Extracts the transcript for the specified language and converts 'start' and 'dur' to floats.

        Parameters:
            tracks (list): List of language blocks with transcripts.
            language (str): The language to extract (e.g., 'English' or 'Hindi (auto-generated)').

        Returns:
            list or None: List of transcript entries with 'start' and 'dur' as floats, or None if not found.
        """
        # Find the matching language block
        transcript = next(
            (track['transcript'] for track in tracks if track['language'].lower() == language.lower()),
            None
        )

        if transcript:
            # Convert 'start' and 'dur' to floats
            for entry in transcript:
                try:
                    entry['start'] = float(entry.get('start', 0))
                except ValueError:
                    entry['start'] = 0.0
                try:
                    entry['dur'] = float(entry.get('dur', 0))
                except ValueError:
                    entry['dur'] = 0.0
            return transcript
        else:
            return None

    def _format_time(self, seconds):
        """
        Converts seconds (float) to MM:SS string format.

        Parameters:
            seconds (float): Time in seconds.

        Returns:
            str: Formatted time string in "MM:SS" format.
        """
        minutes = int(seconds // 60)
        secs = int(seconds % 60)
        return f"{minutes:02}:{secs:02}"

    def generate_meaningful_timestamps(self, transcript):
        load_dotenv()
        GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
        genai.configure(api_key=GOOGLE_API_KEY)

        model = genai.GenerativeModel(
            model_name="gemini-1.5-pro",
            generation_config={
                "temperature": 1,
                "top_p": 0.95,
                "top_k": 64,
                "max_output_tokens": 8192,
                "response_mime_type": "text/plain",
            },
            system_instruction=self.get_instructions('timestream/utils/instructions.txt')
        )

        chat_session = model.start_chat(history=[])

        # Send the transcript to the model
        response = chat_session.send_message("\n".join(transcript))
        cleaned_lines = []
        for line in response.text.splitlines():
            if line.strip():
                # Remove unwanted characters and clean the line
                cleaned_line = line.strip().strip("`").replace('*', '')
                cleaned_lines.append(cleaned_line)
        return cleaned_lines
    


    def user_query(self, transcript, query):
        load_dotenv()
        GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
        genai.configure(api_key=GOOGLE_API_KEY)

        model = genai.GenerativeModel(
            model_name="gemini-1.5-pro",
            generation_config={
                "temperature": 1,
                "top_p": 0.95,
                "top_k": 64,
                "max_output_tokens": 1000,
                "response_mime_type": "text/plain",
            },
            system_instruction=self.get_instructions('timestream/utils/query_instructions.txt')
        )

        chat_session = model.start_chat(history=[])

        # Send the transcript to the model
        try:
            response = chat_session.send_message("\n".join(query) + " ".join(transcript) )
            cleaned_lines = []
            for line in response.text.splitlines():
                if line.strip():
                    # Remove unwanted characters and clean the line
                    cleaned_line = line.strip().strip("`").replace('*', '')
                    cleaned_lines.append(cleaned_line)
            return cleaned_lines
        except:
            response = "The Video is too long for Our model to go through"
            return response
```

```python
# -*- coding: utf-8 -*-

import math

def convert_transcripts(api_response: list, language: str = 'en') -> list:
    """
    Convert only the 'tracks' part of the API response into a
    list of "MM:SS subtitle_text" based on the given language.

    :param api_response: The JSON (as a Python list) from the API.
    :param language: 'en' for English or 'hi' for Hindi (auto-generated).
    :return: A list of timed text strings in the format ['00:00 text', '00:03 text', ...].
    """
    # Map language codes to the label in "tracks"
    language_map = {
        'en': "English",
        'hi': "Hindi (auto-generated)"
    }
    # Default to English if not recognized
    target_language = language_map.get(language, "English")

    final_result = []

    # Go through each item in the response
    for item in api_response:
        # Skip everything except 'tracks'
        tracks = item.get("tracks", [])
        if not tracks:
            continue

        # Find the track that matches our target_language
        selected_transcript = None
        for track in tracks:
            if track.get("language") == target_language:
                selected_transcript = track.get("transcript", [])
                break

        # If this item doesn't have the desired language, move on
        if not selected_transcript:
            continue

        # Build up a list of "MM:SS text" for this item's transcript
        for entry in selected_transcript:
            start_str = entry.get("start", "0")
            text_str = entry.get("text", "")

            # Convert start time to integer seconds, then to MM:SS
            try:
                start_seconds = math.floor(float(start_str))
            except ValueError:
                start_seconds = 0

            minutes = start_seconds // 60
            seconds = start_seconds % 60
            time_formatted = f"{minutes:02d}:{seconds:02d}"

            line = f"{time_formatted} {text_str}"
            final_result.append(line)

    return final_result
```

```python
# -*- coding: utf-8 -*-
import json

def convert_to_subtitle_dicts(
    raw_json_str: str,
    language: str = 'en'
) -> list:
    """
    Convert the raw JSON string (YouTube-like response) into a list of dicts:
    [
      {"text": str, "start": float, "duration": float},
      ...
    ]
    based on the specified language track.

    :param raw_json_str: The entire JSON response as a string.
    :param language: 'en' (English) or 'hi' (Hindi - auto-generated).
    :return: List of transcript entries.
    """
    # Parse the JSON string into Python objects
    data = json.loads(raw_json_str)

    # If it's not a list or empty, return empty
    if not isinstance(data, list) or len(data) == 0:
        return []

    # Extract tracks from the first item
    item = data[0]
    tracks = item.get("tracks", [])

    # Map the user-friendly language code to the actual "language" field
    language_map = {
        'en': "English",
        'hi': "Hindi (auto-generated)"
    }
    target_language = language_map.get(language, "English")

    # Find the transcript matching the chosen language
    selected_transcript = None
    for track in tracks:
        if track.get("language") == target_language:
            selected_transcript = track.get("transcript", [])
            break

    # If no matching track is found, return empty
    if selected_transcript is None:
        return []

    # Build the output structure
    result = []
    for entry in selected_transcript:
        # 'text' is already a string
        text_str = entry.get("text", "")
        
        # Convert start & duration to floats
        try:
            start_val = float(entry.get("start", 0))
        except ValueError:
            start_val = 0.0
        
        try:
            dur_val = float(entry.get("dur", 0))
        except ValueError:
            dur_val = 0.0
        
        # Create the dictionary
        line_dict = {
            "text": text_str,
            "start": start_val,
            "duration": dur_val
        }
        result.append(line_dict)

    return result
```

```python
import re

def extract_transcript(response_tr, language='en'):
    """
    Extracts the transcript for the specified language from the response_tr object.

    Parameters:
        response_tr: An object that has a method get_nlstr() returning the transcript string.
        language (str): The language code to extract ('en' for English, 'hi' for Hindi). Defaults to 'en'.

    Returns:
        list: A list of formatted transcript entries in the "MM:SS text" format for the specified language.
    """

    # Mapping of language codes to language strings used in the transcript data
    language_map = {
        'en': 'english',
        'hi': 'hindi (auto-generated)',
        # Add other languages here if needed
    }

    # Get the full language string based on the language code
    lang_search = language_map.get(language.lower(), 'english')  # Default to 'english'

    # Extract the JSON-like string containing transcripts
    try:
        # Assuming the transcript data is after '"tracks":'
        updated_str = response_tr.get_nlstr().split('"tracks":')[1]
    except IndexError:
        raise ValueError("Invalid response_tr format: 'tracks' key not found.")

    # Regex pattern to capture each language block
    language_block_pattern = r'\{"language":"(.*?)","transcript":\[(.*?)\]\}'

    # Regex pattern to capture text, start, and duration within each transcript
    transcript_pattern = r'\{"text":"(.*?)","start":"(\d+\.?\d*)","dur":"(\d+\.?\d*)"\}'

    # Find all language blocks
    language_blocks = re.findall(language_block_pattern, updated_str, re.DOTALL)

    # Function to convert seconds to MM:SS format
    def seconds_to_mmss(seconds_str):
        try:
            seconds = float(seconds_str)
            minutes = int(seconds // 60)
            secs = int(seconds % 60)
            return f"{minutes:02}:{secs:02}"
        except ValueError:
            return "00:00"  # Default in case of conversion error

    # Dictionary to hold transcripts categorized by language
    transcripts_by_language = {}

    # Process each language block
    for lang, transcripts_str in language_blocks:
        # Find all transcript entries within the current language block
        transcripts = re.findall(transcript_pattern, transcripts_str)

        # Normalize language key to lowercase
        lang_lower = lang.lower()

        # Initialize the list for this language if not already present
        if lang_lower not in transcripts_by_language:
            transcripts_by_language[lang_lower] = []

        # Process each transcript entry
        for text, start, dur in transcripts:
            # Convert start time to MM:SS format
            time_str = seconds_to_mmss(start)
            # Format the transcript entry
            formatted_entry = f"{time_str} {text}"
            # Append to the corresponding language's list
            transcripts_by_language[lang_lower].append(formatted_entry)

    # Retrieve the transcript for the specified language
    specified_transcript = transcripts_by_language.get(lang_search.lower(), [])

    return specified_transcript
```

```python
import re

def convert_subtitles(api_response_str, language='en'):
    """
    Convert the API response string into a list of subtitle lines in "MM:SS text" format.

    This function:
    1. Extracts all complete "text", "start", "dur" entries using regex.
    2. Replaces escaped single quotes with actual single quotes.
    3. Converts the start time from seconds to "MM:SS" format.
    4. Compiles the formatted subtitles into a list.

    :param api_response_str: The raw JSON-like string from the API.
    :param language: Desired language code ('en' or 'hi'). Defaults to 'en'.
    :return: List of formatted subtitle strings.
    """

    # Define a regex pattern to extract "text", "start", and "dur" fields
    # This pattern assumes that "text", "start", and "dur" appear in this order
    pattern = r'"text":"(.*?)","start":"(.*?)","dur":"(.*?)"'

    # Find all matches in the response string
    matches = re.findall(pattern, api_response_str)

    # Initialize the list to hold formatted subtitles
    subtitles = []

    for match in matches:
        text, start, dur = match

        # Replace escaped single quotes with actual single quotes
        text = text.replace("\\'", "'").replace('\\"', '"')

        # Convert start time from string to float
        try:
            start_sec = float(start)
        except ValueError:
            # If conversion fails, default to 0
            start_sec = 0.0

        # Convert seconds to MM:SS format
        minutes = int(start_sec // 60)
        seconds = int(start_sec % 60)
        time_str = f"{minutes:02d}:{seconds:02d}"

        # Combine time and text
        formatted_sub = f"{time_str} {text}"
        subtitles.append(formatted_sub)

    return subtitles

# Example Usage
if __name__ == "__main__":
    # Provided API response string (truncated for brevity)
    api_response_str = response_tr

    # Convert to Hindi subtitles (assuming the provided data is in Hindi)
    # Since the data appears to be in Hindi, set language='hi'
    hindi_subs = convert_subtitles(api_response_str, language="hi")
    print("Hindi Subtitles:")
    for sub in hindi_subs:
        print(sub)
```

```python
import json

def convert_subtitles(api_response_str, language="en"):
    """
    Convert the API response (string) into a list of subtitle lines in "MM:SS text" format.
    
    This function fixes any bracket mismatches and improperly escaped single quotes in the JSON string before parsing.
    
    :param api_response_str: The JSON string you get from the API.
    :param language:        Either 'en' for English or 'hi' for Hindi subtitles.
                            Defaults to 'en'.
    :return:                A list of strings, each containing the time and subtitle text.
    """
    
    def fix_json_string(s):
        """
        Fixes mismatched brackets and improperly escaped single quotes in a JSON string.
        
        :param s: The raw JSON string with potential issues.
        :return:  The corrected JSON string.
        """
        # Step 1: Replace escaped single quotes with actual single quotes
        s = s.replace("\\'", "'")
        
        # Step 2: Remove extra closing brackets from the end
        # Count the number of opening and closing brackets
        opening_brackets = s.count('{') + s.count('[')
        closing_brackets = s.count('}') + s.count(']')
        
        # If there are more closing brackets, remove the extra ones from the end
        while closing_brackets > opening_brackets:
            if s.endswith('}'):
                s = s[:-1]
                closing_brackets -= 1
            elif s.endswith(']'):
                s = s[:-1]
                closing_brackets -= 1
            else:
                # If it doesn't end with a closing bracket, stop
                break
        
        return s
    
    def format_time(seconds_float):
        """
        Converts seconds (float or string) to MM:SS format.
        
        :param seconds_float: Time in seconds (float or string).
        :return:              Time as a string in "MM:SS" format.
        """
        try:
            total_seconds = float(seconds_float)
            minutes = int(total_seconds) // 60
            seconds = int(total_seconds) % 60
            return f"{minutes:02d}:{seconds:02d}"
        except ValueError:
            return "00:00"  # Default fallback
    
    # Step 1: Fix the JSON string
    fixed_json_str = fix_json_string(api_response_str)
    
    # Step 2: Attempt to parse the JSON
    try:
        data = json.loads(fixed_json_str)
    except json.JSONDecodeError as e:
        print(f"JSON decoding failed even after fixing brackets and quotes: {e}")
        return []
    
    # Step 3: Define language mapping
    language_map = {
        "en": "english",
        "hi": "hindi"
    }
    
    desired_language = language_map.get(language.lower())
    
    if not desired_language:
        print(f"Unsupported language code: '{language}'. Supported codes are 'en' and 'hi'.")
        return []
    
    # Initialize the list to hold formatted subtitles
    result_subtitles = []
    
    # Iterate through each language entry in the data
    for entry in data:
        lang_lower = entry.get("language", "").lower()
        
        # Check if the current entry matches the desired language
        if desired_language in lang_lower:
            transcripts = entry.get("transcript", [])
            for item in transcripts:
                start_time = item.get("start", "0")
                text_line = item.get("text", "")
                
                # Format the time
                formatted_time = format_time(start_time)
                
                # Combine time and text
                formatted_line = f"{formatted_time} {text_line}"
                result_subtitles.append(formatted_line)
            break  # Found the desired language, no need to continue
    else:
        print(f"No subtitles found for language '{language}'.")
    
    return result_subtitles

# Example Usage
if __name__ == "__main__":
    # Original API response with an extra closing bracket and improperly escaped single quotes
    api_response_str = response_tr[4].split('"language":')[2].split('"transcript":')[1]
    # Convert to English subtitles
    english_subs = convert_subtitles(api_response_str, language="en")
    print("English Subtitles:")
    for sub in english_subs:
        print(sub)
    
    print("\n" + "-"*50 + "\n")
    
    # Convert to Hindi subtitles
    hindi_subs = convert_subtitles(api_response_str, language="hi")
    print("Hindi Subtitles:")
    for sub in hindi_subs:
        print(sub)
```

```python
import math

def parse_transcript(api_response, language='en'):
    """
    Parse transcript data from the given API response.
    
    Parameters
    ----------
    api_response : list
        The list/JSON structure returned by the YouTube API (or similar).
    language : str, optional
        The language code to filter by, e.g. 'en' or 'hi'. Defaults to 'en'.

    Returns
    -------
    list
        A list of lines in the format ["MM:SS transcript text", "MM:SS transcript text", ...].
        Returns an empty list if no matching language track is found.
    """
    
    # --- 1. Find "tracks" from the response (assuming the response is always a list with at least one element) ---
    if not api_response:
        return []
    
    # Typically the data is the first element in the list, then we read the "tracks" key
    first_item = api_response[0]
    
    tracks = first_item.get('tracks', [])
    if not tracks:
        return []
    
    # --- 2. Identify the track that matches the desired language ---
    #    We compare either the exact languageCode from "languages" or the "language" string in each track.
    #    The structure can vary, so adjust accordingly.
    #    For instance, the sample data has "language" as "English" or "Hindi (auto-generated)",
    #    and "languageCode" as "en" or "hi" in a separate list "languages".
    
    # We'll do a two-step approach:
    #   A) If the track has an exact "languageCode" we can compare that.
    #   B) Else we try to match by "language" string if it contains 'English' or 'Hindi', etc.
    
    # One easy approach: the sample shows two tracks: "English", "Hindi (auto-generated)"
    # We'll match on either:
    #    - "English" if language='en'
    #    - "Hindi" if language='hi'
    
    desired_track = None
    
    # Attempt exact match first on the "languageCode" from the "languages" array:
    # The top-level item has "languages" -> [{"label": "English", "languageCode": "en"}, ...]
    # But each track also has something like track["language"] = "English" or "Hindi (auto-generated)".
    # We'll keep it simple and match the text of track["language"] or "languageCode".
    
    for track in tracks:
        lang_name = (track.get("language") or "").lower()
        
        # If user asked for 'en', check if "english" in lang_name
        # If user asked for 'hi', check if "hindi" in lang_name
        # (You can tweak logic based on how your real data appears.)
        
        if language == 'en' and "english" in lang_name:
            desired_track = track
            break
        if language == 'hi' and "hindi" in lang_name:
            desired_track = track
            break

    # If no matching track was found, return empty list
    if not desired_track:
        return []
    
    # --- 3. Convert each transcript entry to "MM:SS text" ---
    transcript_entries = desired_track.get("transcript", [])
    parsed_list = []
    
    for line in transcript_entries:
        text = line.get("text", "").strip()
        start_str = line.get("start", "0")
        
        # Convert start time to float -> integer
        # e.g. '3.114' -> 3 seconds -> "00:03"
        try:
            start_seconds = float(start_str)
        except ValueError:
            start_seconds = 0
        
        # Floor the seconds to get integer
        start_sec_int = int(math.floor(start_seconds))
        
        # Convert to MM:SS
        mm = start_sec_int // 60
        ss = start_sec_int % 60
        
        # Format "MM:SS"
        timestamp = f"{mm:02d}:{ss:02d}"
        
        # Combine timestamp and text
        parsed_line = f"{timestamp} {text}"
        parsed_list.append(parsed_line)
    
    return parsed_list
```

```python
import json
import math

def convert_subtitles(response, language="en"):
    """
    Convert YouTube transcript data from the API response into a list of strings, 
    where each string is in the format "MM:SS text".

    :param response: The original API response (SList or JSON string).
    :param language: The language code to extract transcripts for ('en' or 'hi').
    :return: A list of subtitle strings in the format ["MM:SS text", ...].
    """

    # If the response is an SList or JSON-like string, parse it into a Python object.
    # If it's already a list/dict, you can skip the json.loads depending on your situation.
    if isinstance(response, str):
        data = json.loads(response)
    else:
        data = response  # Assume it's already a parsed list/dict structure

    # We expect the top-level to be a list with at least one element containing 'tracks'
    if not data:
        return []

    # Extract the "tracks" array from the first item in the list
    tracks = data[0].get("tracks", [])

    # Map language code to the string used in 'language' field
    # Adjust this mapping as needed if the language field naming changes
    target_language = "English" if language == "en" else "Hindi (auto-generated)"

    # Find the track that matches the target language
    for track in tracks:
        if track.get("language") == target_language:
            # Found the desired language track
            transcript_data = track.get("transcript", [])
            formatted_subs = []

            for entry in transcript_data:
                text = entry.get("text", "")
                start = entry.get("start", "0")

                # Convert start to float
                try:
                    start_float = float(start)
                except ValueError:
                    start_float = 0.0

                # Convert to mm:ss format
                minutes = int(start_float // 60)
                seconds = int(start_float % 60)
                timestamp = f"{minutes:02d}:{seconds:02d}"

                formatted_subs.append(f"{timestamp} {text}")

            return formatted_subs

    # If no matching language track was found, return an empty list
    return []
```
